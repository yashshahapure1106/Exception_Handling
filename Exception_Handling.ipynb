### Handling above errors with try except handling

```python
def division_exc() :
    try:
        num = float(input('Please enter numerator  : '))
        den = float(input('Please enter denominator  :' ))
        div = num/den
        return div
    except ZeroDivisionError:
        print('Denominator cannot be 0 in division')
    except ValueError:
        print('Please enter numbers only, alphabets are not allowed for division')
```

```python
print(division_exc())
print('Calculation Complete')
```

```python
print(division_exc())
print('Calculation Complete')
```

```python
print(division_exc())
print('Calculation Complete')
```

### Try Except Finally block

1. Code will execute try block
2. if any error/exception occurs it will go to except block
3. finally block will always execute

```python
def calculate_square_root():
    try:
        n = float(input('Please Enter a number: '))
        import math
        s = math.sqrt
        s = math.sqrt(n)
        return s
    except Exception as e:
        print(f'Exception occured : {e}')
    finally:
        print('Finally Block is always executed')
```

```python
calculate_square_root()
```

```python
calculate_square_root()
```
